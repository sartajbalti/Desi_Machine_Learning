### ***Dataset X includes the following features:***

- f1: Ticket Purchase Date Time
- f2: Origin
- f3: Destination
- f4: Departure Date Time
- f5: Arrival Date Time
- f6: Airline
- f7: Refundable Ticket
- f8: Baggage Weight
- f9: Baggage Pieces
- f10: Flight Number

```python
# Changing the string into number we consider origin as 0 and Destination as 1
df = df.replace('x',0)
```

```python
# Changing the string into number we consider origin as 0 and Destination as 1
df = df.replace('y',1)
```

```python
#df.head()
```

```python
# mean max method on f8 baggage weight to normalize the data
df['f8'] = (df['f8']-df['f8'].min())/(df['f8'].max()-df['f8'].min())
```

```python
#df.head()
```

```python
#Replacing the Booleans to 1/0
df["f7"] = df["f7"].astype(int)
```
